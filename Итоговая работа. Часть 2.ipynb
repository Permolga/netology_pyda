```python
import pandas as pd
```

```python
columns = ['User Id', 'Movie Id', 'Rating', 'Timestamp']
data = pd.read_csv('u.data.csv', sep = '\t', header = None, names = columns)
```

```python
data.head()
```

```python
columns2 = ['Movie Id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('u.item.csv', sep='|', names=columns2,
encoding='latin-1')
```

```python
items.head()
```

```python
#Пользователь, поставивший больше всего оценок
data_user = data.groupby('User Id') ['Rating'].count().sort_values(ascending=False)
```

```python
data_user.head()
```

```python
#Оставим только фильмы, которые смотрел №405
data_405 = data[(data['User Id'] == 405)]
```

```python
data_405.head()
```

```python
#Посчитаем количество рейтингов для каждого фильма
stats = data.groupby('Movie Id')['Rating'].count()
```

```python
stats.head(20)
```

```python
#А сейчас посчитаем среднее значение рейтинга
stats2 = data.groupby('Movie Id')['Rating'].sum()
```

```python
stats2.head()
```

```python
#Добавляем признаки и статистику, для этого объединяем таблицу активного пользователя и фильмы
merged1 = data_405.merge(items, on='Movie Id', how = 'inner')
```

```python
merged1.head()
```

```python
#Уберем лишнее
merged1 = merged1[['Movie Id', 'Rating', 'release date' ,'unknown', 'Action', 'Adventure',
'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
```

```python
merged1.head()
```

```python
#Добавим теперь статистику
merged2 = merged1.merge(stats, on='Movie Id', how = 'inner').merge(stats2, on='Movie Id', how = 'inner')
```

```python
merged2.head()
```

```python
#Заменим дату на год
merged2['release date'] = pd.to_datetime(merged2['release date'], format='%d-%b-%Y')
```

```python
merged2['year']=pd.DatetimeIndex(merged2['release date']).year
```

```python
merged3 = merged2[['Movie Id', 'Rating_x','unknown', 'Action', 'Adventure',
'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'Rating_y', 'Rating', 'year']]
```

```python
merged3.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Формируем целевую переменную
X, y = merged3 [['Movie Id', 'Rating','Rating_y','unknown', 'Action', 'Adventure',
'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'year']], merged3['Rating_x']
```

```python
#Делим данные
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# Приводим данные к одному масштабу
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
X_train.head()
```

```python
#Построим модель линейной регрессии
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X_train, y_train)
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
print('Slope: {:.2f}'.format(slr.coef_[0]))
print('Intercept: {:.2f}'.format(slr.intercept_))
```

```python
#Проверим качество
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score

print('MSE train: {:.3f}, test: {:.3f}'.format(
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: {:.3f}, test: {:.3f}'.format(
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```
