5.С использованием t-test статистики проверить взаимосвязи между несколькими переменными.

### Количество проектов и количество часов в месяц

```python
#Выберем тех, кто выполнял 3 и больше проекта в месяц, и тех, кто выполнял меньше трех, 
#выведем также среднее количество часов в месяц
more_hours = data[data['number_project'] >= 3] ['average_montly_hours']
less_hours = data[data['number_project'] < 3] ['average_montly_hours']
len(more_hours), len(less_hours)
```

```python
#Выбираем по 50 образцов из каждой выборки
more_hours = more_hours.sample(50)
less_hours = less_hours.sample(50)
```

```python
#Смотрим, нормально ли распределены данные
statistics.variance(more_hours), statistics.variance(less_hours)
```

```python
#Делаем T-test с распределением Велча
sts.ttest_ind(more_hours, less_hours, equal_var=False)
```

p-value меньше 0,05, получается, мы можем отклонить нулевую гипотезу

### Уровень удовлетворенности и зарплата

```python
#Выберем людей с высокой и низкой зарплатой, а также посмотрим их уровень удовлетворенности
happy_low = data[data['salary'] == 'low'] ['satisfaction_level']
happy_high = data[data['salary'] == 'high'] ['satisfaction_level']
len(happy_low), len(happy_high)
```

```python
#Выбираем по 50 образцов из каждой выборки
happy_low = more_hours.sample(50)
happy_high = less_hours.sample(50)
```

```python
#Смотрим, нормально ли распределены данные
statistics.variance(happy_low), statistics.variance(happy_high)
```

```python
#Делаем T-test с распределением Велча
sts.ttest_ind(happy_low, happy_high, equal_var=False)
```

p-value меньше 0,05, получается, мы можем отклонить нулевую гипотезу

10.Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию.
