```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')
data.head()
```

```python
def class_rating(cl):
    """Определяет уровень рейтинга"""
    for i in data:
        if cl <= 2.0:
            return ('Low')
        elif cl <=4.0:
            return ('Middle')
        else:
            return ('High')
```

```python
data['class'] = data['rating'].apply(class_rating)
data.head()
```

```python
data2 = pd.read_csv('keywords.csv')
data2.head()
```

```python

geo_data = {

'Центр': ['москва', 'тула', 'ярославль', 'вк', 'страница'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def region(a):    
    if a in geo_data.get('Центр'):
        return 'Центр'
    else:
        return 'undefined'
```

```python
data2['region'] = data2['keyword'].apply(region)
data2.head(10)
```
