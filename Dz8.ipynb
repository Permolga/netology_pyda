Задание 1

```python
ratings = pd.read_csv('ratings.csv')
ratings_grouped = ratings.groupby('userId').count()
ratings_filtered = ratings_grouped[ratings_grouped['rating'] >=100].reset_index()
film_fans_user_ids = ratings_filtered['userId'].tolist()
active =ratings [ratings['userId'].isin(film_fans_user_ids)]
active_grouped = active.groupby('userId').agg(['min', 'max'])['timestamp'].reset_index()
def time_difference (a):
    return (a['max']-a['min'])
active_grouped ['time_difference'] = active_grouped.apply(time_difference, axis=1)
active_grouped['time_difference'].mean()
```
