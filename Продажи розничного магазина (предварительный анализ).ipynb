```python
import pandas as pd
```

```python
data = pd.read_csv('Online Retail.csv')
data.head()
```

```python
data.info()
```

```python
# Тип поля стоимости одной единицы продукции не является числом. Для того, чтобы его преобразовать в число с плавающей точкой,
# сначала заменим запятую на точку, а затем сделаем преобразование.
data['UnitPrice'] = [x.replace(',', '.') for x in data['UnitPrice']]
data['UnitPrice'] = data['UnitPrice'].astype(float)
```

```python
#Теперь стоимость одной единицы продукции - это число, и можно посчитать сумму строки.Добавим для этого новое поле Total
data['Total'] = (data['Quantity']*data['UnitPrice'])
data.head()
```

```python
#Сгруппируем данные по чекам (InvoiceNo), посмотрим общие суммы чеков
data_grouped_inv = data.groupby('InvoiceNo').sum()
data_grouped_inv
```

```python
#Посмотрим количество покупок по странам
data_grouped_countries = data.groupby('Country').sum().sort_values('Country').reset_index()
data_grouped_countries
```

```python
#Выберем только Финляндию
data_fin = data [(data['Country'] == 'Finland')]
data_fin.head(20)
```

```python
#Сгруппируем по чекам
data_invoices = data.groupby('InvoiceNo').sum()
data_invoices
```

```python
#Посчитаем среднюю сумму чека
av_inv = (data_invoices ['Total'].sum())/(data_invoices ['Total'].count())
av_inv
```

```python
#Посчитаем, сколько товаров в среднем в чеке
av_num = (data_invoices ['Quantity'].sum())/(data_invoices ['Quantity'].count())
av_num
```
