```python
import pandas as pd
```

```python
#Это данные о рождении щенков всех пород собак в Финляндии с 2014 по 2018 год. Собаки делятся на группы. Например, группа №9
#это декоративные собаки.
data = pd.read_csv('dogs4.csv')
data.head()
```

```python
#Группируем данные о рождении щенков по группам ФЦИ, считаем сумму в каждой группе в каждом году
data_grouped = data.groupby('FCI group').sum().sort_values('FCI group').reset_index()
data_grouped
```

```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 12,9

# импортируем нужные библиотеки
import matplotlib.pyplot as plt
```

```python
#Сравниваем, сколько собак родилось в 2018 году в каждой группе ФЦИ, с помощью столбчатой диаграммы
plt.bar(data_grouped['FCI group'], data_grouped['2018'])
plt.xlabel('Group')
plt.ylabel('Quantity')
plt.grid()
```

Больше всего собак родилось в 5-ой группе, причем разница с остальными очень заметная. Меньше всего собак - в 10-й группе. По количеству выделяются также 9-я, 1-я и 8-я группы.

```python
#Находим Top-5 пород по количеству щенков в 9 группе в 2018 году
data_group9 = data[(data['FCI group'] == 'FCI 9')].sort_values('2018', ascending=False)
data_group9_2018 = data_group9[['FCI group', 'Breed','2018']].head(5)
data_group9_2018
```

```python
data_group9_2018.plot(kind='pie', y='2018')
```

Разница между количеством собак видна плохо, хотя синий кусочек пирога все-таки выделяется, и заметно, что фиолетовый - самый маленький. По остальным сделать выводы сложно.

```python
#Посмотрим, как менялось количество щенков разновидностей пуделя по годам
data_poodles = data[((data['Breed'] == 'Miniature Poodles') |  (data['Breed'] == 'Toy Poodles') | (data['Breed'] == 'Medium Size Poodles'))]
data_poodles_new = data_poodles[['Breed','2014', '2015', '2016', '2017', '2018']]
data_poodles_new
```

```python
#Посмотрим, как менялось количество щенков пуделей всех разновидностей за 5 лет
data_poodles_new.plot(kind='bar')
```

Видим, что миниатюрных пуделей рождается в целом больше, чем других разновидностей. Меньше всего тоев. Количество щенков в целом за 5 лет уменьшилось во всех трех породах. Наибольшая разница опять же в миниатюрной разновидности, наименьшая - у малых. Количество щенков той пуделя за последние три года почти одинаковое, а в 2018 году их родилось немного больше, чем в предыдущем. Малых родилось меньше всего в 2017 году, в 2018 их количество снова возросло.

```python
#Группируем данные о рождении щенков по годам, считаем сумму в каждой группе в каждом году и общее количество, находим Top-10
data_breeds = data.groupby('Breed').sum().sort_values('2018', ascending=False).reset_index().head(11)
data_breeds_2018 = data_breeds[['Breed','2018']].head(11)
data_breeds_2018
```

```python
import seaborn as sns
```

```python
#Импортируем данные питомников больших пуделей. В таблице название питомника, количество пометов, щенков, год первого помета и последнего
data = pd.read_csv('StandardPoodles.csv')
data
```

```python
#Выделяем питомники, которым 10 и больше лет
data2 = data[(data['Latest litter'] - data['First litter']) >=10].filter(items = ['Kennel','Offspring', 'Litters'])
data2
```

```python
#Посмотрим связь между количеством пометов и щенков в питомниках.
sns_plot = sns.pairplot(
    data2[['Offspring', 'Litters']])
sns_plot
```

Не очень хорошо понимаю эти графики, если честно, но на них видно, что количество щенков явно прямо коррелирует с количеством пометов. Есть один питомник, который очень интенсивно производит щенков. Интересно, что есть разрывы на графиках, т.е. не бывает значений примерно во втором квартиле, количество либо совсем небольшое, либо выше среднего.
