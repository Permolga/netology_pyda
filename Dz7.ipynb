```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')
data.head()
```

```python
def class_rating(cl):
    """Определяет уровень рейтинга"""
    for i in data:
        if cl <= 2.0:
            return ('Low')
        elif cl <=4.0:
            return ('Middle')
        else:
            return ('High')
```

```python
data['class'] = data['rating'].apply(class_rating)
data.head()
```

```python
data2 = pd.read_csv('keywords.csv')
data2.head()
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def region(a):    
    for k,v in geo_data.items():
        for word in v:
            if word in str(a):
                return k
        else:
            return 'undefined'
    
```

```python
data2['region'] = data2['keyword'].apply(region)
data2.head(50)
```

```python
data3 = pd.read_csv('movies.csv')
data3.head()
```

```python
year = range(1950,2011)
```

```python
def year_production (y):
    for g in year:
        if str(g) in str(y):
            return int(g)
    else:
        return 1900
```

```python
data3['year'] = data3['title'].apply(year_production)
data3.head()
```

```python
data_new=pd.merge(data, data3, on='movieId', how='left') [['title','rating','year']]
data_new.head()
```

```python
data_grouped = data_new.groupby('year').mean().sort_values('rating', ascending=False)
data_grouped.head()
```
